```python
print("CALCULATING CIRCUMFRENCE and AREA OF A CIRCLE")
p= 3.14
print("Radius of the circle is...")
r= float(input("Enter the radius of your circle in centimetres" ))
print(f"{r} cm")
print("After calculation...")
d= float(r*2)
circumfrence=  p*d 
area= p*pow(r,2)
print(f"The circumfrence of your circle is {circumfrence} cm")
print(f"The area of your circle is {area} cm")
```
